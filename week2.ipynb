```python
import numpy as np
import pandas as pd

df = pd.read_csv('restaurants.csv')
```

```python
cheap_restaurants = (df['Price Range'] == '$') | (df['Price Range'] == '$$')
print(cheap_restaurants)
```

```python
good_restaurants = (df['Rating (%)'] >= 80) & (df['Rating (%)'] <= 90)
print(good_restaurants)
```

```python
datetime = pd.to_datetime(df['Opening Date'])
thursday_openings = (datetime.dt.month == 3)
print(thursday_openings)
```
